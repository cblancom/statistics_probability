```python
import numpy as np 
import matplotlib.pyplot as plt

def probabilidad_dado_frecuentista(num_caras, num_ensayos, cara_objetivo):
    """
    Código que calcula la propiedad frecuentista del experimentoq ue consiste
    en lanzar un dado N veces.


    param num_caras (int): Número de caras del dado.
    param num_ensayos (int): Número de ensayos del experimento.
    param cara_objetivo (int): Cara de interés en el experimento.

    return (numpy array): Arreglo con la probabilidad calculada en cada ensayo.
    """
    
    probs = np.zeros((num_ensayos,))
    caras_obtenidas = 0
    for i in range(num_ensayos):
      cara = np.random.randint(1,num_caras+1)
      if cara == cara_objetivo:
          caras_obtenidas += 1    
      probs[i] = caras_obtenidas / (i+1)  
    
    return probs



def plot_probs(probs, num_caras, hline=False):
  """
  Código para graficar un array con probabilidades conseguidas en mumtiples
  ensayos de un experimento. 

  param_probs (numpy array): Arreglo con la probabilidad calculada en cada 
                             ensayo.
  param_num_caras (int): Número de caras del dado usado en el experimento.
  param hline (bool): Variable que muestra la probabilidad según la definición
                      clásica.          
  """

  plt.figure(dpi=100)
  plt.plot(probs, label=('$F_A(n)$'))
  if hline:
    plt.axhline(1/num_caras, color='red', label='1/'+str(num_caras))
  plt.title('Probabilidad de obtener un {} al lanzar un dado de {} caras.'.format(cara_objetivo, num_caras))
  plt.xlabel('Número de lanzamientos')
  plt.ylabel('Probabilidad')
  plt.legend()
  plt.grid()
  plt.show()  
```

```python
num_ensayos = 100000
cara_objetivo = 4
num_caras = 6

probs = probabilidad_dado_frecuentista(num_caras, num_ensayos, cara_objetivo)
plot_probs(probs, num_caras, hline=True)
```
